Foi utilizado os dados da cidade de Florianópolis, já que a cidade de Correa Pinto não apresenta dados no site do prof, e a cidade mais próxima com dados seria Florianópolis.

```python
# Selecionar arquivo para análise
import os

# Pegando o caminho do diretório que estou
rootPath = os.getcwd()
print(rootPath)

# Definindo o diretório com os arquivos metar
metSuperficiePath = rootPath +'\\inputs\\meteoSuperficie'
print(metSuperficiePath)

# Listando os arquivos dentro do diretório
files = os.listdir(metSuperficiePath)
print(files)
```

```python
# Abrindo os arquivos metar
import pandas as pd

# Abrindo cada arquivo dentro da pasta e acumulando em uma lista chamada de metSuperficie
metSuperficie=[]
for file in files:
    metSuperficie.append(pd.read_csv(metSuperficiePath+'\\'+file))

# Convertendo metSuperficie para DataFrame
metSuperficie = pd.concat(metSuperficie)

print(metSuperficie)
print(type(metSuperficie))
```

```python
# Criar uma coluna com Datetime dos dados

# Criar uma coluna de data e hora combinando as colunas separadas
metSuperficie['datetime'] = pd.to_datetime(metSuperficie[['year', 'month', 'day', 'hour']])
print(metSuperficie['datetime'])

# Definido a coluna de datetime como índice da matriz
metSuperficie.set_index(metSuperficie['datetime'])
```

```python
# Analisando os dados através de figuras
import matplotlib.pyplot as plt

fig,ax = plt.subplots(6,sharex=True)
ax[0].plot(metSuperficie['datetime'],metSuperficie['temp'],color='purple')
ax[1].plot(metSuperficie['datetime'],metSuperficie['p'],color='violet')
ax[2].plot(metSuperficie['datetime'],metSuperficie['relh'])
ax[3].plot(metSuperficie['datetime'],metSuperficie['cco'],color='gray')
#ws velocidade
ax[4].plot(metSuperficie['datetime'],metSuperficie['ws'],color='pink')
#wd direção
ax[5].plot(metSuperficie['datetime'],metSuperficie['wd'],color='yellow')
ax[5].tick_params(axis='x', rotation=45)
```

```python
# Fazendo boxplot das variáveis
fig,ax = plt.subplots(1,6,sharex=True)
ax[0].boxplot(metSuperficie['temp'])
ax[1].boxplot(metSuperficie['p'])
ax[2].boxplot(metSuperficie['relh'])
ax[3].boxplot(metSuperficie['cco'])
ax[4].boxplot(metSuperficie['ws'])
ax[5].boxplot(metSuperficie['wd'])

# Removendo valores acima do percentil 95°
import numpy as np
temp99 = np.percentile(metSuperficie['temp'],99)
print('O percentil 99 da temperatura é:'+str(temp99))
p99 = np.percentile(metSuperficie['p'],99)
print('O percentil 99 da pressão é:'+str(p99))
#print(metSuperficie['temp']<temp99)
cco99 = np.nanpercentile(metSuperficie['cco'],99)
print('O percentil 99 da cco é:'+str(p99))
ws99 = np.nanpercentile(metSuperficie['ws'],99)
print('O percentil 99 da ws é:'+str(ws99))
wd99 = np.nanpercentile(metSuperficie['wd'],99)
print('O percentil 99 da wd é:'+str(wd99))

fig,ax = plt.subplots(1,6,sharex=True)
ax[0].boxplot(metSuperficie['temp'][(metSuperficie['temp']<temp99) & 
              (metSuperficie['temp']>np.percentile(metSuperficie['temp'],0.5))])
ax[1].boxplot(metSuperficie['p'][(metSuperficie['p']<p99) & 
              (metSuperficie['p']>np.percentile(metSuperficie['p'],0.5))])
ax[2].boxplot(metSuperficie['relh'])
ax[3].boxplot(metSuperficie['cco'][~np.isnan(metSuperficie['cco'])])
ax[4].boxplot(metSuperficie['ws'])
ax[5].boxplot(metSuperficie['wd']) #achei estranho ficar sem dado no gráfico
```

```python
# Estatísticas dos dados
print(metSuperficie)
metSuperficie.set_index('datetime', inplace=True)
media_mensal = metSuperficie.resample('M').mean()
min_mensal = metSuperficie.resample('M').min()

# Plotando figura após fazer a média mensal
fig,ax = plt.subplots(6,sharex=True)
ax[0].plot(media_mensal.index,media_mensal['temp'],color='purple')
ax[1].plot(media_mensal.index,media_mensal['p'],color='violet')
ax[2].plot(media_mensal.index,media_mensal['relh'])
ax[3].plot(media_mensal.index,media_mensal['cco'],color='gray')
#ws velocidade
ax[4].plot(media_mensal.index,media_mensal['ws'],color='pink')
#wd direção
ax[5].plot(media_mensal.index,media_mensal['wd'],color='yellow')
ax[5].tick_params(axis='x', rotation=45)

```

```python
media_mensal = metSuperficie.resample('M').mean()
min_mensal = metSuperficie.resample('M').min()
max_mensal = metSuperficie.resample('M').max()

# Plotando as estatísticas com sombra
fig, ax = plt.subplots(6, sharex=True, figsize=(15, 20))

# Plot para a temperatura
ax[0].plot(media_mensal.index, media_mensal['temp'], color='red')
ax[0].fill_between(media_mensal.index, min_mensal['temp'], max_mensal['temp'], color='red', alpha=0.2)
ax[0].set_ylabel('Temp')

# Plot para a pressão
ax[1].plot(media_mensal.index, media_mensal['p'], color='orange')
ax[1].fill_between(media_mensal.index, min_mensal['p'], max_mensal['p'], color='orange', alpha=0.2)
ax[1].set_ylabel('Pressure')

# Plot para a umidade relativa
ax[2].plot(media_mensal.index, media_mensal['relh'], color='navy')
ax[2].fill_between(media_mensal.index, min_mensal['relh'], max_mensal['relh'], color='navy', alpha=0.2)
ax[2].set_ylabel('Relative Humidity')

# Plot para a cobertura de nuvens
ax[3].plot(media_mensal.index, media_mensal['cco'], color='gray')
ax[3].fill_between(media_mensal.index, min_mensal['cco'], max_mensal['cco'], color='gray', alpha=0.2)
ax[3].tick_params(axis='x', rotation=45)
ax[3].set_ylabel('Cloud Cover')

# Plot para a velocidade do vento
ax[4].plot(media_mensal.index, media_mensal['ws'], color='springgreen')
ax[4].fill_between(media_mensal.index, min_mensal['ws'], max_mensal['ws'], color='springgreen', alpha=0.2)
ax[4].set_ylabel('Wind Speed')

# Plot para a direção do vento
ax[5].plot(media_mensal.index, media_mensal['wd'], color='dodgerblue')
ax[5].fill_between(media_mensal.index, min_mensal['wd'], max_mensal['wd'], color='dodgerblue', alpha=0.2)
ax[5].set_ylabel('Wind Direction')

plt.show()
```
