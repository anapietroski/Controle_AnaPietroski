```python
# Selecionar arquivo para análise
import os

# Pegando o caminho do diretório que estou
rootPath = os.getcwd()
print(rootPath)

# Definindo o diretório com os arquivos metar
metSuperficiePath = rootPath +'\\inputs\\meteoSuperficie'
print(metSuperficiePath)

# Listando os arquivos dentro do diretório
files = os.listdir(metSuperficiePath)
print(files)
```

```python
# Abrindo os arquivos metar
import pandas as pd

# Abrindo cada arquivo dentro da pasta e acumulando em uma lista chamada de metSuperficie
metSuperficie=[]
for file in files:
    metSuperficie.append(pd.read_csv(metSuperficiePath+'\\'+file))

# Convertendo metSuperficie para DataFrame
metSuperficie = pd.concat(metSuperficie)

print(metSuperficie)
print(type(metSuperficie))
```

```python
# Criar uma coluna com Datetime dos dados

# Criar uma coluna de data e hora combinando as colunas separadas
metSuperficie['datetime'] = pd.to_datetime(metSuperficie[['year', 'month', 'day', 'hour']])
print(metSuperficie['datetime'])

# Definido a coluna de datetime como índice da matriz
metSuperficie.set_index(metSuperficie['datetime'])
```

```python
# Importação de bibliotecas necessárias
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from siphon.simplewebservice.wyoming import WyomingUpperAir
import matplotlib.pyplot as plt

# Definição das datas e estação
# Definir datas de verão e inverno
summer_dates = [datetime(2023, 1, 10) + timedelta(days=i) for i in range(5)]
winter_dates = [datetime(2023, 7, 10) + timedelta(days=i) for i in range(5)]

# Estação de monitoramento no aeroporto de Florianópolis
station = 'SBFL'

# Função para extrair dados de temperatura
def extract_temperature_profile(dates, station):
    profiles = {'0h': [], '12h': []}
    for date in dates:
        for hour in [0, 12]:
            try:
                # Extrair dados
                data = WyomingUpperAir.request_data(date.replace(hour=hour), station)
                # Adicionar dados ao perfil correspondente
                profiles[f'{hour}h'].append(data)
            except Exception as e:
                print(f"Erro ao extrair dados para {date} {hour}h: {e}")
    return profiles

# Extrair dados para as datas definidas
# Extrair perfis de temperatura para verão e inverno
summer_profiles = extract_temperature_profile(summer_dates, station)
winter_profiles = extract_temperature_profile(winter_dates, station)

# Análise do perfil de temperatura
def analyze_temperature_profile(profiles):
    inversions = {'0h': [], '12h': []}
    boundary_layers = {'0h': [], '12h': []}
    
    for time in ['0h', '12h']:
        for profile in profiles[time]:
            # Extrair altura e temperatura
            height = profile['height']
            temperature = profile['temperature']
            
            # Determinar inversão
            inversion = None
            for i in range(1, len(temperature)):
                if temperature[i] > temperature[i-1]:
                    inversion = height[i]
                    break
            inversions[time].append(inversion)
            
            # Determinar altura da camada limite (simplificada)
            boundary_layer_height = height[np.argmax(np.gradient(temperature))]
            boundary_layers[time].append(boundary_layer_height)
    
    return inversions, boundary_layers

# Analisar e plotar os resultados
def plot_temperature_profiles(profiles, title):
    fig, ax = plt.subplots(1, 2, figsize=(14, 8))
    times = ['0h', '12h']
    colors = ['blue', 'orange']
    
    for i, time in enumerate(times):
        for profile in profiles[time]:
            ax[i].plot(profile['temperature'], profile['height'], color=colors[i], alpha=0.5)
        ax[i].set_xlabel('Temperature (°C)')
        ax[i].set_ylabel('Height (m)')
        ax[i].set_title(f'{title} {time}')
        ax[i].invert_yaxis()
    
    plt.show()

# Plotar perfis de temperatura
plot_temperature_profiles(summer_profiles, 'Verão')
plot_temperature_profiles(winter_profiles, 'Inverno')

# Analisar perfis de temperatura
summer_inversions, summer_boundary_layers = analyze_temperature_profile(summer_profiles)
winter_inversions, winter_boundary_layers = analyze_temperature_profile(winter_profiles)

print("Verão - Inversões (0h):", summer_inversions['0h'])
print("Verão - Inversões (12h):", summer_inversions['12h'])
print("Verão - Camadas Limite (0h):", summer_boundary_layers['0h'])
print("Verão - Camadas Limite (12h):", summer_boundary_layers['12h'])
print("Inverno - Inversões (0h):", winter_inversions['0h'])
print("Inverno - Inversões (12h):", winter_inversions['12h'])
print("Inverno - Camadas Limite (0h):", winter_boundary_layers['0h'])
print("Inverno - Camadas Limite (12h):", winter_boundary_layers['12h'])


```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from siphon.simplewebservice.wyoming import WyomingUpperAir
import matplotlib.pyplot as plt


# Definir datas de verão e inverno
summer_dates = [datetime(2023, 1, 10) + timedelta(days=i) for i in range(5)]
winter_dates = [datetime(2023, 7, 10) + timedelta(days=i) for i in range(5)]

# Estação de monitoramento no aeroporto de Florianópolis
station = 'SBFL'


def extract_temperature_profile(dates, station):
    profiles = []
    for date in dates:
        daily_profile = {}
        for hour in [0, 12]:
            try:
                # Extrair dados
                data = WyomingUpperAir.request_data(date.replace(hour=hour), station)
                daily_profile[f'{hour}h'] = data
            except Exception as e:
                print(f"Erro ao extrair dados para {date} {hour}h: {e}")
        profiles.append(daily_profile)
    return profiles


def extract_temperature_profile(dates, station):
    profiles = []
    for date in dates:
        daily_profile = {}
        for hour in [0, 12]:
            try:
                # Extrair dados
                data = WyomingUpperAir.request_data(date.replace(hour=hour), station)
                daily_profile[f'{hour}h'] = data
            except Exception as e:
                print(f"Erro ao extrair dados para {date} {hour}h: {e}")
        profiles.append(daily_profile)
    return profiles



# Extrair perfis de temperatura para verão e inverno
summer_profiles = extract_temperature_profile(summer_dates, station)
winter_profiles = extract_temperature_profile(winter_dates, station)



def plot_daily_temperature_profiles(profiles, title, dates):
    fig, axs = plt.subplots(len(profiles), 2, figsize=(14, 4*len(profiles)), sharex=True, sharey=True)
    colors = ['blue', 'orange']
    
    for i, daily_profile in enumerate(profiles):
        for j, time in enumerate(['0h', '12h']):
            if time in daily_profile:
                data = daily_profile[time]
                axs[i, j].plot(data['temperature'], data['height'], color=colors[j], alpha=0.7)
                axs[i, j].set_xlabel('Temperature (°C)')
                if j == 0:
                    axs[i, j].set_ylabel(f'Height (m)\n{dates[i].strftime("%Y-%m-%d")}')
                axs[i, j].set_title(f'{title} {time}')
                axs[i, j].invert_yaxis()
    
    plt.tight_layout()
    plt.show()

# Plotar perfis de temperatura
plot_daily_temperature_profiles(summer_profiles, 'Verão', summer_dates)
plot_daily_temperature_profiles(winter_profiles, 'Inverno', winter_dates)

def analyze_temperature_profile(profiles):
    inversions = []
    boundary_layers = []
    
    for daily_profile in profiles:
        daily_inversions = {}
        daily_boundary_layers = {}
        
        for time in ['0h', '12h']:
            if time in daily_profile:
                data = daily_profile[time]
                height = data['height']
                temperature = data['temperature']
                
                # Determinar inversão
                inversion = None
                for i in range(1, len(temperature)):
                    if temperature[i] > temperature[i-1]:
                        inversion = height[i]
                        break
                daily_inversions[time] = inversion
                
                # Determinar altura da camada limite (simplificada)
                boundary_layer_height = height[np.argmax(np.gradient(temperature))]
                daily_boundary_layers[time] = boundary_layer_height
        
        inversions.append(daily_inversions)
        boundary_layers.append(daily_boundary_layers)
    
    return inversions, boundary_layers


# Analisar perfis de temperatura
summer_inversions, summer_boundary_layers = analyze_temperature_profile(summer_profiles)
winter_inversions, winter_boundary_layers = analyze_temperature_profile(winter_profiles)

print("Verão - Inversões:", summer_inversions)
print("Verão - Camadas Limite:", summer_boundary_layers)
print("Inverno - Inversões:", winter_inversions)
print("Inverno - Camadas Limite:", winter_boundary_layers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from meteostat import Point, Daily
from datetime import datetime

# Definindo a estação (coordenadas do aeroporto de Florianópolis)
fln = Point(-27.6703, -48.5525)

# Definindo o período de interesse
start = datetime(2023, 1, 1)
end = datetime(2023, 1, 31)

# Obtendo dados diários de temperatura
data = Daily(fln, start, end)
data = data.fetch()

# Verificando os dados obtidos
print(data.head())

# Plotando o perfil de temperatura
plt.figure(figsize=(10, 5))
plt.plot(data.index, data['tavg'], marker='o', linestyle='-')
plt.title('Perfil de Temperatura - Aeroporto de Florianópolis (SBFL)')
plt.xlabel('Data')
plt.ylabel('Temperatura Média (°C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
