# Aula 04 - revisando comandos Python

```python
# Criando uma lista em python
list = [0, 'eu amo polAtm', 2.5]
print(list)

# Nesta parte fiz a adição entre duas strings, sendo list[1] é o segundo indice do vetor list
print('variável: '+ list[1])

# Adição do primeiro indice da variável list (list[0]) - Lembrar que a indexação no python inicia em 0
print(list[0] + 1)
```

```python
# Criando uma matriz no python
import numpy as np

# Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print(a)
print ('Indexação numpy')
print(a[0,0])
```

```python
# Criando números aleatórios e mostrando a dimensão da matriz
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
# Criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 # valor mínimo
vmax = 100 # valor máximo

# Matriz com valores aleatórios dentro de uma faixa estabelecida
c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz 
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
# Plotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
plt.title('Gráfico com Valores Interiros no Eixo X')
plt.xticks(x) # Definindo os valores do eixo x
```

```python
# Criando um gráfico de barras e exportando 

fig, ax =  plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# Exportar a figura 
import os
rootFolder = os.getcwd()
print(rootFolder)
fig.savefig('Figure01.png')
```
