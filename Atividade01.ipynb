```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Coletando o caminho da pasta de que estou
cwd = os.getcwd()

# Pasta dos fatores de emiss√£o
path = os.path.join(cwd, 'tabelas', 'FE')

# Indicando nome do arquivo
chapterNumer = 10
subChapter = 2
file = 'chp_'+str(chapterNumer)+'_'+str(subChapter)+'.csv'

# Abrindo arquivo utilizando o Pandas
fe = pd.read_csv(os.path.join(path, file))

# Definindo algumas vari√°veis
n = 365 # 365 dias
atvMax = 0.00225 # atividade m√°xima de produ√ß√£o em ton/s
atvMin = 0.00150 # atividade m√≠nima di√°ria de produ√ß√£o em ton/s

dias = np.arange(1, n+1)  # Dias

# Lista de poluentes
poluentes = ['MP', 'SO2', 'CO']

# Para cada fonte de emiss√£o, vamos criar as tabelas de emiss√£o
for source in fe['Source'].unique():
    # Criando um DataFrame para armazenar os resultados
    resultados = pd.DataFrame(columns=['Dia'] + poluentes)
    
    # Para cada dia, vamos calcular as emiss√µes e armazenar no DataFrame
    for dia in dias:
        # Calculando a sazonalidade senoidal
        sazonalidade = atvMax - (atvMax - atvMin) * np.sin(2 * np.pi * dia / n)
        
        # Gerando valores aleat√≥rios com sazonalidade senoidal para cada poluente
        aleatorioSazonal = (atvMax - atvMin) * np.random.rand(len(poluentes)) + atvMin + sazonalidade
        
        # Calculando as emiss√µes em g/s para cada poluente
        emissao_gs = aleatorioSazonal * 1000 / (60 * 60)  # Convertendo de ton/s para g/s
        
        # Criando um dicion√°rio com os dados do dia e as emiss√µes de cada poluente
        dados_dia = {'Dia': dia}
        for i, poluente in enumerate(poluentes):
            dados_dia[poluente] = emissao_gs[i]
        
        # Adicionando os dados do dia ao DataFrame
        resultados.loc[len(resultados)] = dados_dia
    
    # Salvando a tabela de emiss√£o em um arquivo CSV
    resultados.to_csv(os.path.join(path, 'emis_'+source+'.csv'), index=False)

    # Exibindo a tabela de emiss√£o
    print(f'Tabela de emiss√£o para {source}:')
    print(resultados)
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Definindo os fatores de emiss√£o por processo (fonte) e poluente
path = os.path.join(cwd, 'tabelas', 'FE')  # Substitua pelo caminho correto do seu arquivo

# Definindo a taxa de atividade em toneladas/ano
atv = 70000

# Calculando as emiss√µes para cada poluente e fonte
emissoes_por_processo = fe.drop(columns='Source') * atv

# Calculando a soma das emiss√µes considerando todos os processos
soma_emissoes = emissoes_por_processo.sum().sum()  # Soma de todas as c√©lulas do DataFrame

# Calculando a emiss√£o m√©dia, m√≠nima e m√°xima
emissao_media = emissoes_por_processo.mean().mean()
emissao_minima = emissoes_por_processo.min().min()
emissao_maxima = emissoes_por_processo.max().max()

# Exibindo os resultados
print("Soma das emiss√µes de todos os processos:", soma_emissoes, "g/s")
print("Emiss√£o m√©dia:", emissao_media, "g/s")
print("Emiss√£o m√≠nima:", emissao_minima, "g/s")
print("Emiss√£o m√°xima:", emissao_maxima, "g/s")

# Plotando os gr√°ficos para cada poluente
for poluente in emissoes_por_processo.columns:
    # Obtendo os dados de emiss√£o para o poluente atual
    dados_emissao = emissoes_por_processo[poluente]
    
    # Plotando o gr√°fico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(dados_emissao.index, dados_emissao.values, color='skyblue')
    plt.title(f'Emiss√£o de {poluente} por fonte', fontsize=14)
    plt.xlabel('Fonte', fontsize=12)
    plt.ylabel('Emiss√£o (g/s)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
